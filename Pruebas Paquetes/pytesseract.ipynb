Podemos obtener las bounding boxes del texto:

```python
from PIL import ImageDraw, ImageFont
import matplotlib.pyplot as plt


# Cargar imagen
img = Image.open(imgs/"fox.jpg")
draw = ImageDraw.Draw(img)

# Obtener cajas de Tesseract
boxes = pytesseract.image_to_boxes(img)

# Dibujar cajas
for b in boxes.splitlines():
    b = b.split()
    char, x1, y1, x2, y2, _ = b
    x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
    # PIL tiene origen en la esquina superior izquierda, Tesseract en la inferior
    y1_new = img.height - y2
    y2_new = img.height - y1
    draw.rectangle([x1, y1_new, x2, y2_new], outline="red")
    font = ImageFont.truetype("arial.ttf", size=20)
    draw.text((x1, y1_new - 10), char, fill="blue", font=font)

# Mostrar imagen
plt.imshow(img)
plt.axis('off')
plt.show()
```

Bueno se pueden ver algunos guesses de "-" por elementos del entorno, a ver la confianza de esos guesses:

```python
from pathlib import Path
from PIL import Image
import pytesseract

img_path = Path("imgs") / "fox.jpg"
img = Image.open(img_path)

data = pytesseract.image_to_data(img, lang="spa")

for line in data.splitlines()[1:]:
    parts = line.split('\t')
    if len(parts) == 12:
        x, y, w, h = map(int, parts[6:10])
        try:
            conf = int(parts[10])
        except ValueError:
            conf = -1  # si está vacío o no es número
        word = parts[11]
        print(word, conf, x, y, w, h)
```
