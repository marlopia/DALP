# Ejercicio 0

Hemos hecho un estudio para averiguar en qué invierten su tiempo libre las personas con edades comprendidas entre 30 – 60 años.
Las repuestas que hemos recibido son las siguientes e indican el nº de horas que dedican al día a:
Escuchar música: 3, 1, 6, 2, 4, 0, 5, 3, 8, 1, 2

**Instrucciones:** Calcula las siguientes estadísticas usando **SOLO operaciones con listas** (sin pandas ni numpy):
- Media
- Mediana (valor central después de ordenar los datos)
- Moda (valor que más se repite)
- Desviación típica
- Mínimo
- Máximo

```python
# Lista de horas dedicadas a escuchar música
lst_musica = [3, 1, 6, 2, 4, 0, 5, 3, 8, 1, 2]

# TODO: Ordena la lista

lst_musica.sort()


# TODO: Calcula el mínimo (usa la función min() o encuentra el valor más pequeño)

print("min: ", min(lst_musica))

# TODO: Calcula el máximo (usa la función max() o encuentra el valor más grande)

print("max: ", max(lst_musica))

# TODO: Calcula la media (suma de todos los valores / cantidad de valores)

print("media: ", sum(lst_musica) / len(lst_musica))

# TODO: Calcula la mediana (el valor en la posición central de la lista ordenada)
# Pista: Si la lista tiene n elementos, la mediana está en la posición n//2
print("mediana:",lst_musica[len(lst_musica)//2])

# TODO: Calcula la moda (el valor que más se repite)
# Pista: Puedes usar el método count() para contar cuántas veces aparece cada valor
# O puedes usar collections.Counter si prefieres
print("moda:",max(set(lst_musica), key=lst_musica.count))

# TODO: Calcula la desviación típica
# Pista: desviación típica = raíz cuadrada de la varianza
# varianza = suma de (cada_valor - media)^2 / cantidad_de_valores
# Necesitarás: import math y usar math.sqrt()
import math
print("desviacion: ",math.sqrt((sum((x - sum(lst_musica)/len(lst_musica))**2 for x in lst_musica) / len(lst_musica))))


# TODO: Imprime los resultados
```

# Ejercicio 1

Hemos hecho un estudio para averiguar en qué invierten su tiempo libre las personas con edades comprendidas entre 30 – 60 años.
Las repuestas que hemos recibido son las siguientes e indican el nº de horas que dedican al día a:
Escuchar música: 3, 1, 6, 2, 4, 0, 5, 3, 8, 1, 2
Jugar a videojuegos: 2, 5, 1, 3, 0, 6, 2, 1, 4, 3, 0
Ver la TV: 4, 7, 2, 5, 3, 8, 4, 6, 2, 3, 1
Practicar deporte: 1, 4, 2, 3, 5, 2, 6, 1, 3, 2, 0

**Instrucciones:** Usa pandas para cargar los datos en un DataFrame y calcular las estadísticas descriptivas.

```python
import pandas as pd

# TODO: Crea 4 listas con los datos de cada actividad del enunciado
# lst_musica = ...
# lst_videojuegos = ...
# lst_tv = ...
# lst_deporte = ...
lst_musica = [3, 1, 6, 2, 4, 0, 5, 3, 8, 1, 2] 
lst_videojuegos = [2, 5, 1, 3, 0, 6, 2, 1, 4, 3, 0]
lst_tv = [4, 7, 2, 5, 3, 8, 4, 6, 2, 3, 1]
lst_deporte = [1, 4, 2, 3, 5, 2, 6, 1, 3, 2, 0]

# TODO: Crea un DataFrame con todas las actividades
# Usa pd.DataFrame() con un diccionario donde las claves sean los nombres de las columnas
# y los valores sean las listas correspondientes

data = {
    "musica": lst_musica,
    "videojuegos": lst_videojuegos,
    "tv": lst_tv,
    "deporte": lst_deporte
}

df = pd.DataFrame(data)

# TODO: Muestra el DataFrame creado

print(df)


# TODO: Calcula las estadísticas descriptivas para cada actividad
# Puedes usar un bucle for para recorrer cada columna del DataFrame
# Para cada columna, calcula:
#   - Media: df[columna].mean()
#   - Mediana: df[columna].median()
#   - Moda: df[columna].mode()[0]
#   - Desviación típica: df[columna].std()
#   - Mínimo: df[columna].min()
#   - Máximo: df[columna].max()

for col in df.columns.tolist():
    slice = df[col]
    print("Estadisticas para ",col)
    print(f"Media {col}: {slice.mean()}")
    print(f"Mediana {col}: {slice.median()}")
    print(f"Moda {col}: {slice.mode()[0]}")
    print(f"Desviación típica {col}: {slice.std()}")
    print(f"Mínimo {col}: {slice.min()}")
    print(f"Máximo {col}: {slice.max()}")


# TODO: Usa el método describe() para obtener un resumen estadístico completo
df.describe()
```
