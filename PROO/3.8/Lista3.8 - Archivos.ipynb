## EJERCICIO 7

Crea una lista de diccionarios con información de 3 libros (Título, Autor, Año). Escribe esta lista en un archivo llamado `libros.csv` utilizando `csv.DictWriter`.

````python
libro1 = {
    "titulo": "Alicia en el País de las Maravillas",
    "autor": "Lewis Carroll",
    "año": "1865",
}

libro2 = {
    "titulo": "1984",
    "autor": "George Orwell",
    "año": "1949",
}

libro3 = {
    "titulo": "Fahrenheit 451",
    "autor": "Ray Bradbury",
    "año": "1953",
}

with open("data/libros.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["titulo", "autor", "año"])
    writer.writeheader()
    writer.writerow(libro1)
    writer.writerow(libro2)
    writer.writerow(libro3)
````

## EJERCICIO 9

Crea un diccionario con la configuración de una aplicación (ej: tema, idioma, notificaciones). Guarda este diccionario en un archivo `config.json`. Luego, lee el archivo y muestra la configuración cargada.

{
  "app_config": {
    "theme": "dark",
    "language": "es",
    "notifications": {
      "enabled": true,
      "sound": true,
      "vibration": false,
      "daily_summary": true
    },
    "privacy": {
      "share_usage_data": false,
      "allow_personalization": true
    },
    "user_preferences": {
      "font_size": "medium",
      "auto_save": true,
      "backup_frequency": "weekly"
    },
    "version": "1.0.3"

````python
import json

data = {
  "app_config": {
    "theme": "dark",
    "language": "es",
    "notifications": {
      "enabled": True,
      "sound": True,
      "vibration": False,
      "daily_summary": True
    },
    "privacy": {
      "share_usage_data": False,
      "allow_personalization": True
    },
    "user_preferences": {
      "font_size": "medium",
      "auto_save": True,
      "backup_frequency": "weekly"
    },
    "version": "1.0.3"
  }
}

with open("data/config.json", "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=4)

with open("data/config.json", "r", encoding="utf-8") as f:
    data_read = json.load(f)

print(data_read)
````

## EJERCICIO 11

Investiga sobre el módulo `pickle` para la serialización de objetos.
1. Define una clase `Persona` con atributos `nombre` y `edad`.
2. Crea una instancia de dicha clase.
3. Guarda el objeto en un archivo binario llamado `persona.pkl`.
4. Lee el archivo y recupera el objeto original.

````python
import pickle

class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def __str__(self):
        return f"Persona(nombre={self.nombre}, edad={self.edad})"

persona = Persona("Alba", 30)

with open("data/persona.pkl", "wb") as f:
    pickle.dump(persona, f)

with open("data/persona.pkl", "rb") as f:
    persona_recuperada = pickle.load(f)

print(persona_recuperada)
````
