## Ejercicio 2: Indexing y Slicing

Dado este array:

```python
arr = np.array([10, 20, 30, 40, 50, 60, 70, 80])
print(f"Array: {arr}")
```

2.1) Obtén el primer elemento

```python
# Tu código aquí
arr[0]
```

2.2) Obtén el último elemento

```python
# Tu código aquí
arr[-1]
```

2.3) Obtén los elementos del índice 2 al 5 (inclusive)

```python
# Tu código aquí
arr[1:5]
```

2.4) Obtén cada segundo elemento (índices 0, 2, 4, 6...)

```python
# Tu código aquí
for i in range(1,len(arr)+1):
    if not i % 2:
        print(arr[i-1])
```

## Ejercicio 5: Casos prácticos

5.1) **Aumento de salarios:** 
- Array de salarios actuales: [2000, 2500, 3000, 1800, 2200]
- Calcula los nuevos salarios con un aumento del 10%

```python
salarios_actuales = np.array([2000, 2500, 3000, 1800, 2200])

# Tu código aquí
# Incrementa 10% (multiplica por 1.10)
nuevos_salarios = salarios_actuales * 1.1
print(nuevos_salarios)
```

5.2) **Conversión de temperatura:**
- Array de temperaturas en Celsius: [0, 15, 30, 45]
- Convierte a Fahrenheit usando la fórmula: F = C * 9/5 + 32

```python
celsius = np.array([0, 15, 30, 45])

# Tu código aquí

fahrenheit = celsius * (9/5) + 32
print(fahrenheit)
```

5.3) **Selección de valores altos:**
- Array de calificaciones: [65, 78, 92, 55, 88, 76]
- Cuenta cuántas calificaciones son mayores a 80

```python
calificaciones = np.array([65, 78, 92, 55, 88, 76])

# Tu código aquí

mayores_ochenta = calificaciones[calificaciones > 80]
print(len(mayores_ochenta))
```
