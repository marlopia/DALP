## Ejercicio 6: Casos prácticos

6.1) **Dataset de salarios:**
- Crea una Series con salarios: [2000, 2500, 3000, 1800]
- Calcula: total, promedio, mínimo, máximo

```python
# Tu código aquí
salarios = pd.Series([2000, 2500, 3000, 1800])

print("Total", salarios.sum())
print("Promedio", salarios.mean())
print("Máximo", salarios.max())
print("Mínimo", salarios.min())
```

6.2) **Calificaciones:**
- Crea una Series: [85, 90, 78, 92, 88]
- Cuenta cuántas están por encima de 85

```python
# Tu código aquí
calificaciones = pd.Series([85, 90, 78, 92, 88])

print(len(calificaciones[calificaciones > 85]))
```

6.3) **Aumento de sueldo:**
- Series de sueldos actuales
- Calcula nuevos sueldos con aumento del 10%

```python
# Tu código aquí
salarios = pd.Series([2000, 2500, 3000, 1800])

nuevos_salarios = salarios * 1.1

print(nuevos_salarios)
```
