#  Nivel 5‚Äì6 ¬∑ Outliers y an√°lisis cr√≠tico

##  Ejercicio 5

Se registraron las **ventas diarias** de una tienda a lo largo de un **mes**. Hubo **un d√≠a con ventas muy altas** respecto al resto.

## Tareas
- Detectar **posibles outliers**.
- Calcular **media** y **mediana** **con y sin** ese valor.
- Explicar **qu√© pudo pasar** ese d√≠a (rebajas, campa√±a, error‚Ä¶).

üéØ Storytelling: "**Un solo d√≠a excepcional** distorsiona la **percepci√≥n** del mes completo".

```python
import numpy as np
import plotly.express as px

ventas = np.random.uniform(100, 1000, 30)

ventas = np.append(ventas, 2500)

df = pd.DataFrame({"Ventas": ventas})

Q1 = df["Ventas"].quantile(0.25)
Q3 = df["Ventas"].quantile(0.75)
IQR = Q3 - Q1

lim_inf = Q1 - 1.5 * IQR
lim_sup = Q3 + 1.5 * IQR

df["Outlier"] = (df["Ventas"] < lim_inf) | (df["Ventas"] > lim_sup)

print("Outliers detectados:")
print(df[df["Outlier"]])

print("Media con outlier: ", df["Ventas"].mean())
print("Mediana con outlier: ", df["Ventas"].median())

df_limpio = df[(df["Ventas"] >= lim_inf) & (df["Ventas"] <= lim_sup)]

print("Media sin outlier: ", df_limpio["Ventas"].mean())
print("Mediana sin outlier: ", df_limpio["Ventas"].median())

fig = px.box(df, y="Ventas", points="outliers", title="Boxplot de Ventas")
fig.show()
```

## Ejercicio 7: Notas por trimestre


Se tienen las **notas del mismo grupo** en **3 trimestres**.

trimestre1 = [5.5, 6.0, 5.8, 6.2, 7.0, 6.5, 5.9, 6.1, 7.2, 6.8, 5.4, 6.0, 6.3]

trimestre2 = [6.2, 6.5, 6.8, 7.1, 7.3, 6.9, 7.0, 6.7, 7.5, 7.4, 6.6, 7.2, 7.1]

trimestre3 = [7.0, 7.2, 7.5, 7.8, 8.0, 7.9, 7.6, 7.7, 8.2, 8.1, 7.4, 7.3, 7.9]

## Tareas
- Crear un **diagrama de caja** por trimestre.
- Comparar **mediana** y **rango intercuart√≠lico (IQR)** de cada trimestre.
- Explicar la **evoluci√≥n del grupo**.

üéØ Storytelling: "No solo sube la **nota central**, tambi√©n se **reduce la dispersi√≥n**‚Ä¶"

```python
import plotly.express as px

trimestre1 = [5.5, 6.0, 5.8, 6.2, 7.0, 6.5, 5.9, 6.1, 7.2, 6.8, 5.4, 6.0, 6.3]
trimestre2 = [6.2, 6.5, 6.8, 7.1, 7.3, 6.9, 7.0, 6.7, 7.5, 7.4, 6.6, 7.2, 7.1]
trimestre3 = [7.0, 7.2, 7.5, 7.8, 8.0, 7.9, 7.6, 7.7, 8.2, 8.1, 7.4, 7.3, 7.9]

data = pd.DataFrame({
    "Trimestre 1": trimestre1,
    "Trimestre 2": trimestre2,
    "Trimestre 3": trimestre3
})

# Convertir a formato largo
df_long = data.melt(var_name="Trimestre", value_name="Calificaci√≥n")

fig = px.box(df_long, x="Trimestre", y="Calificaci√≥n", title="Boxplot por Trimestre")
fig.show()

for col in data.columns:
    q1 = data[col].quantile(0.25)
    q3 = data[col].quantile(0.75)
    iqr = q3 - q1
    mediana = data[col].median()
    print(f"{col} ‚Üí Mediana: {mediana}, IQR: {iqr}")
```

Se ve una evoluci√≥n claramente positiva dado a que la mediana aumenta y el IRQ se reduce, por tanto hay m√°s beneficios de manera m√°s consistente

### Comparaci√≥n con y sin outliers
